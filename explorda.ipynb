```python
def plot_line_graph(x, y, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
def plot_scatter_graph(x, y, title, xlabel='0', ylabel='1'):
    plt.figure(figsize=(10, 9))
    plt.scatter(x, y, c='b')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
def plot_histogram(x, bins, title, xlabel='0', ylabel='1'):
    plt.figure(figsize=(10, 9))
    plt.hist(x, bins)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
def plot_bar(x, y, title, xlabel, ylabel):
    plt.figure(figsize=(10,9))
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
# plot_line_graph(df['total_views'], df['country'], 'Views vs Country', 'Views', 'Country')
plot_bar(df_country['country'], df_country['subscribers'], 'country vs subscribers', 'country', 'subsribers')
plot_bar(df_country['country'], df_country['total_views'], 'country vs total_views', 'country', 'total_views')
plot_bar(df_country['country'], df_country['total_videos'], 'country vs total_videos', 'country', 'total_videos')

# plot_scatter_graph(df['channel_name'], df['total_views'], 'channel_name vs channel_views', 'channel_name', 'total_views')
# plot_histogram(df['total_views'], 1000, "Histogram distribution of channels by total views")
```

```python
# Update dataset with latest data from youtube

from pyyoutube import Api
import json
import time
from tqdm import tqdm
```

```python
DATATYPE_VIDEOS = "video"
DATATYPE_CHANNEL = "channel"
DATATYPES = [DATATYPE_VIDEOS, DATATYPE_CHANNEL]

def get_items_batch(api: Api, datatype, arr, batch_size=50):
    print(f"Processing {datatype} data")
    all_items = []
    for i in tqdm(range(0, len(arr), batch_size)):
        batch = arr[i:i + batch_size]
        # print(f"Processing batch {i}")
        if datatype == DATATYPE_VIDEOS:
            response = api.get_video_by_id(video_id=batch)
        elif datatype == DATATYPE_CHANNEL:
            response = api.get_channel_info(channel_id=batch)
            
        parsed = json.loads(response.to_json())
        normalized = pd.json_normalize(parsed['items'])
        
        normalized['scrape_date'] = time.time_ns()
        
        all_items.append(normalized)
    
    # Concat all DataFrames into one
    return pd.concat(all_items, ignore_index=True)


def extract_dataset(t, arr):
    if t not in DATATYPES:
        return [0]
    
    with open('api_key.json', 'r') as f:
        creds = json.load(f)

    api = Api(api_key=creds['api_key'])
    
    df = get_items_batch(api, t, arr)
    df.columns = [field.lower().replace(".", "_") for field in df]
    
    save_to_file(t, df)

    return df

def save_to_file(t, df: pd.DataFrame, path="data/youtube_tech_channels/processed/"):
    epoch = time.time_ns()
    df.to_csv(path + f"youtube_tech_{t}_{epoch}.csv")

videos_new = extract_dataset(DATATYPE_VIDEOS, df_videos['video_id'].to_list())
channels_new = extract_dataset(DATATYPE_CHANNEL, df['channel_id'].to_list())
```

```python
channels_new.rename(columns={"id": "channel_id"}, inplace=True)
videos_new.rename(columns={"channel_id": "video_id"}, inplace=True)
channels_new.shape, videos_new.shape
```

```python
# clean data

def preprocessor(df: pd.DataFrame, threshold_percentage: float) -> pd.DataFrame:
    # drop empty columns larger than threshold
    # merge and then drop duplicates
    missing = df.isnull().sum() / len(df)
    to_drop = missing[missing > threshold_percentage].index.tolist()
    print(f"Dropped {len(to_drop)}")
    new_df = df.drop(columns=to_drop, inplace=False)
    return new_df

channels_new = preprocessor(channels_new, 0)
videos_new = preprocessor(videos_new, 0)

channel_merge = df.merge(channels_new, "outer", "channel_id")
video_merge = df_videos.merge(videos_new, "outer", "video_id")

# identify and remove duplicate columns

channels_dup_cols = ['channel_name', 'description', 'subscribers', 'total_views', 'total_videos', 'uploads_playlist', 'scraped_at', 'kind', 'etag']
videos_dup_cols = ['title', 'published_at', 'video_url', 'snippet_channeltitle','scraped_at', 'kind', 'etag']

channel_merge = channel_merge.drop(columns=channels_dup_cols, inplace=False)
video_merge = video_merge.drop(columns=videos_dup_cols, inplace=False)
```
