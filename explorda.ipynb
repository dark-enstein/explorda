```python
df_country = df_merge[['country','total_views','total_videos','subscribers']].groupby('country', as_index=False).sum()
df_country
```

```python
def plot_line_graph(x, y, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
def plot_scatter_graph(x, y, title, xlabel='0', ylabel='1'):
    plt.figure(figsize=(10, 9))
    plt.scatter(x, y, c='b')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
def plot_histogram(x, bins, title, xlabel='0', ylabel='1'):
    plt.figure(figsize=(10, 9))
    plt.hist(x, bins)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
def plot_bar(x, y, title, xlabel, ylabel):
    plt.figure(figsize=(10,9))
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
    
# plot_line_graph(df['total_views'], df['country'], 'Views vs Country', 'Views', 'Country')
plot_bar(df_country['country'], df_country['subscribers'], 'country vs subscribers', 'country', 'subsribers')
plot_bar(df_country['country'], df_country['total_views'], 'country vs total_views', 'country', 'total_views')
plot_bar(df_country['country'], df_country['total_videos'], 'country vs total_videos', 'country', 'total_videos')

# plot_scatter_graph(df['channel_name'], df['total_views'], 'channel_name vs channel_views', 'channel_name', 'total_views')
# plot_histogram(df['total_views'], 1000, "Histogram distribution of channels by total views")
```
